```python
#import libraries pandas and sqlalchemy
import pandas as pd
from sqlalchemy import create_engine , text 
```

```python
#we have to read master data cleaned 

df=pd.read_csv('demo/bestofrt_master.csv')
```

```python
df.head()
```

```python
#create engine with database empty form bestofrt.db
engine=create_engine('sqlite:///demo/bestofrt.db')
```

```python
#using to_sql func to store data from df dataframe into bestofrt.db by engine
df.to_sql('rt_table',engine,index=False,if_exists='replace')
```

```python
#read database 
df_gather=pd.read_sql(sql=text('select * from rt_table'),con=engine.connect())
df_gather
```
